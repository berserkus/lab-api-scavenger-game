# Challenge 1: Fork Languages

You will find out how many programming languages are used among all the forks created from the main lab repo of your bootcamp. Assuming the main lab repo is ironhack-datalabs/datamad1020-rev, you will:

1. Obtain the full list of forks created from the main lab repo via Github API.

2. Loop the JSON response to find out the language attribute of each fork. Use an array to store the language attributes of each fork.

Hint: Each language should appear only once in your array.

3. Print the language array. It should be something like:

["Python", "Jupyter Notebook", "HTML"]


**HINT:**: Create a list with every **language_url** you find in every fork 

⚠️ **Remember to limit your output before submission**.

```python
user="ironhack-datalabs"
repo="datamad1020-rev"
url=f'https://api.github.com/repos/{user}/{repo}/forks'
```

```python
url_lang=f'https://api.github.com/repos/{user}/{repo}/languages'
url_lang
```

```python
parameters={'Authorizationf':f'token {github}'}
```

```python
parameters
```

```python
res=requests.get(url, params=parameters) #call the api to receive the info
```

```python
df=pd.DataFrame(res.json()) #store it into a dataframe
```

```python
df
```

```python
list_forks=list(df['id'])
```

```python
languages_from_df=list(df['language'])
```

```python
url_list=list(df['url']) #store the urls of all fork
```

```python
languages_from_df
```

```python
# what I thought you are looking for
fork_languages=list(zip(list_forks,languages_from_df))
fork_languages
```

```python
# what you are actually looking for

# create a loop that will loop through the database
languages=[]
for i in url_list:
    url_test=i
    url_test=url_test+"/languages"
    res_test=requests.get(url_test, params=parameters)
    languages.append(res_test.json())
    
```

```python
languages
```

```python
fork_languages_2=list(zip(list_forks,languages))
fork_languages_2
```
